Handling Missing Values

```python
print(df.isnull().sum())
```

```python
df.fillna('0', inplace=True)
```

Data Cleaning

```python
# Dropping irrelevant columns

df.drop(columns=['Employee_Name', 'EmpID'], inplace=True)
```

```python
# Convert date columns to datetime format

df['DOB'] = pd.to_datetime(df['DOB'])
df['DateofHire'] = pd.to_datetime(df['DateofHire'])
df['LastPerformanceReview_Date'] = pd.to_datetime(df['LastPerformanceReview_Date'])
```

```python
# Check for duplicates and remove if necessary

df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.dtypes
```

Feature Engineering

```python
# Create tenure column

df['DateofTermination'] = pd.to_datetime(df['DateofTermination'], errors='coerce') # Use errors='coerce' to handle invalid dates
df['DateofHire'] = pd.to_datetime(df['DateofHire'], errors='coerce') # Use errors='coerce' to handle invalid dates

df['Tenure'] = (df['DateofTermination'].fillna(pd.Timestamp('today')) - df['DateofHire']).dt.days / 365
```

```python
# Calculate age

df['Age'] = (pd.Timestamp('today') - df['DOB']).dt.days / 365
```

```python
# Convert terminated status to binary

df['Termd'] = df['Termd'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
# Tenure and Career Progression

# Create a 'Tenure' column by calculating the difference between Date of Hire and Date of Termination
df['Tenure'] = (pd.to_datetime(df['DateofTermination']) - pd.to_datetime(df['DateofHire'])).dt.days / 365

# Scatter plot of Tenure vs Performance
plt.figure(figsize=(10,6))
sns.scatterplot(x='Tenure', y='PerformanceScore', data=df)
plt.title('Tenure vs Performance Score')
plt.xlabel('Tenure (Years)')
plt.ylabel('Performance Score')
plt.show()
```

```python
# Recruitment and Performance

# Convert 'PerformanceScore' to numeric values if it contains strings

# Check the unique values in 'PerformanceScore' column
print(df['PerformanceScore'].unique())

# If it contains strings like 'Needs Improvement', 'Fully Meets', etc.,
# create a mapping dictionary and convert them to numeric scores.

performance_mapping = {'Needs Improvement': 1, 'Fully Meets': 2, 'Exceeds': 3}
df['PerformanceScore'] = df['PerformanceScore'].map(performance_mapping)

# calculate the average performance score by recruitment source
recruitment_performance = df.groupby('RecruitmentSource')['PerformanceScore'].mean()

# Plot
plt.figure(figsize=(10,6))
recruitment_performance.plot(kind='bar')
plt.title('Average Performance Score by Recruitment Source')
plt.xlabel('Recruitment Source')
plt.ylabel('Average Performance Score')
plt.xticks(rotation=90)
plt.show()
```

```python
# Engagement Survey Analysis

# Scatter plot of Engagement Survey vs Satisfaction
plt.figure(figsize=(10,6))
sns.scatterplot(x='EngagementSurvey', y='EmpSatisfaction', data=df)
plt.title('Engagement Survey Score vs Employee Satisfaction')
plt.xlabel('Engagement Survey Score')
plt.ylabel('Employee Satisfaction')
plt.show()
```

```python
# Manager Impact

# Group by Manager and average Performance Score

manager_performance = df.groupby('ManagerName')['PerformanceScore'].mean()

# Plot
plt.figure(figsize=(10,6))
manager_performance.plot(kind='bar')
plt.title('Average Performance Score by Manager')
plt.xlabel('Manager')
plt.ylabel('Average Performance Score')
plt.xticks(rotation=90)
plt.show()
```
